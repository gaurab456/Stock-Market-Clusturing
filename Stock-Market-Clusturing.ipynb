```python
comapanies = {'Tata Motors Ltd':'TTM','Infosys Limited':'INFY','Cognizant Technology Solutions':'CTSH',
              'ICICI Bank Limited':'IBN','HDFC Bank Limited':'HDB','Wipro Limited':'WIT','Amazon':'AMZN','Apple':'AAPL','Walgreen':'WBA',
              'Boeing':'BA','Lockheed Martin':'LMT','McDonalds':'MCD','Intel':'INTC','Navistar':'NAV','IBM':'IBM','MasterCard':'MA','Microsoft':'MSFT','Pepsi':'PEP','Honda':'HMC',
            'Sony':'SNE','Exxon':'XOM'}
```

```python
data_source = 'yahoo'
start_date = '2015-04-25'
end_date = date.today()
df = data.DataReader(list(comapanies.values()),data_source,start_date,end_date)
```

```python
#Difference of opening and closing stock prices for a particular day. Positive movement implies we can buy the stock. 
movements = stock_close-stock_open
```

```python
Sum_movements = np.sum(movements,axis=1)
```

As we can see MacD and Amazon have different scales. There is a significant difference in rate of chnge of units in stock
price and in that case we need to normalize our data in order to have consistency in our data for further andnalysis. 


```python
from sklearn.preprocessing import Normalizer
Normalizer = Normalizer()
norm_movements = Normalizer.fit_transform(movements)
```

```python
norm_movements
```

```python
plt.figure(figsize=(28,10))

plt.subplot(1,2,1)
plt.title('MacDonalds Stock Movement',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Date',fontsize=15)
plt.ylabel('Movement',fontsize=15)
plt.plot(norm_movements[11])

plt.subplot(1,2,2)
plt.title('Amazon Stock Movement',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Date',fontsize=15)
plt.ylabel('Movement',fontsize=15)
plt.plot(norm_movements[6])
```

```python
#Model Building
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline

Normalizer = Normalizer()
kmeans = KMeans(n_clusters=3,max_iter=800, random_state=2)


pipeline = make_pipeline(Normalizer,kmeans)
pipeline.fit(movements)
predictions = pipeline.predict(movements)
```

```python
predictions
```

```python
clusters = pd.DataFrame({'companies':list(comapanies),'Cluster':predictions})
clusters
```

There are 1434 of features, we can use PCA to reduce the dimension to visualize our data


```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import Normalizer

Normalizer = Normalizer()
data_pca = PCA(n_components=2)
kmeans = KMeans(n_clusters=3,max_iter=800, random_state=2)
pipeline = make_pipeline(Normalizer,data_pca,kmeans)
pipeline.fit(movements)
predictions = pipeline.predict(movements)
```

```python
clusters2 = pd.DataFrame({'companies':list(comapanies),'Cluster':predictions}).sort_values(by=['Cluster'])
clusters2
```

```python
#Visualizing decision boundary
import matplotlib.pyplot as plt

reduced_data = PCA(n_components=2).fit_transform(norm_movements)
kmeans = KMeans(init="k-means++", n_clusters=3, n_init=4)
kmeans.fit(reduced_data)

# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .02     # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation="nearest",
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired, aspect="auto", origin="lower")

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=169, linewidths=3,
            color="w", zorder=10)
plt.title("K-means clustering on stock data (PCA-reduced data)\n"
          "Centroids are marked with white cross")
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
